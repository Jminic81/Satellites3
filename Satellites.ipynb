```python
#Analysis and Visualization of Active Satellites in Orbit around Earth; Jonathan Bouchet's Kaggle submission.  
#The Readme contains original background and discussion of the data analysis in this Jupyter Notebook. 

#Import all libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings  #Chose to ignore the warnings that are printed by the Seaborn library
warnings.filterwarnings("ignore")
sns.set(style="white", color_codes=True)
```

```python
#Read the first .csv file into data frame "sat".
sat=pd.read_csv("database.csv")
```

```python
#Summarize the location and number of cells that are blank.  See Readme for discussion of cells that are not blank but do not
#contain valid data.
sat.isnull().sum()
```

```python
#Summarize Class of Orbit in descending order.
sat["Class of Orbit"].value_counts()
```

```python
#Use bar chart (pareto chart) to show Class of Orbit by count in descending order.
Class_of_Orbit = sat['Class of Orbit'].value_counts()
Class_of_Orbit.head(4).plot(kind='bar')
plt.title('Satellite Class of Orbit')
plt.xlabel('Orbit Class')
plt.ylabel('Number of Satellites')
plt.show()
```

```python
#Summarize Dry Mass (Kilograms) in descending order.
sat["Dry Mass (Kilograms)"].value_counts()
```

```python
#Summarize Country of Operator/Owner in descending order.
sat["Country of Operator/Owner"].value_counts()
```

```python
#Use bar chart (pareto chart) to show Country of Operator/Owner by count in descending order.
Country_Operator_Owner = sat["Country of Operator/Owner"].value_counts()
Country_Operator_Owner.head(20).plot(kind='bar')
plt.title('Satellite Ownership by Count')
plt.xlabel('Country of Operator/Owner')
plt.ylabel('Number of Satellites')
plt.show()
```

```python
#Summarize Purpose in descending order.
sat["Purpose"].value_counts()
```

```python
#Use pie chart to summarize purpose - best to only use 6 "slices" so bottom 12 items were combined into "Other" 
labels = ['Communications', 'Earth Observation', 'Technology Development', 'Navigation/GPS', 'Earth/Space Science', 'Other']
sizes = [699,355,154,97,71,43]

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels = labels, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.title('Satellite Purpose')
plt.show()
```
